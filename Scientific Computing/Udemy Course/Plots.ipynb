Generating Data Points

```python
#data points
#(starting value, end value, number of points between)
x1 = np.linspace(0,4,101)
x1
```

```python
y1 = 1 + 2*np.cos(3*x1)
y1
```

```python
plt.xlim([0,8])
plt.plot(x1,y1)
```

```python
x2 = np.linspace(4,8,101)
y2 = 1 + 2*np.cos(3*x2)
```

```python
plt.xlim([0,8])
plt.plot(x1,y1)
plt.plot(x2,y2)
```

```python
x2, y2 = np.meshgrid(np.linspace(-10,10,201),np.linspace(-10,10,201)
)
z2 = x2 + y2**2
```

```python
contourplot = plt.contour(x2,y2,z2)
plt.clabel(contourplot, inline=1, fontsize=10)
# Labels indicate value of z: on "0" line all values are the same, "15" on 15 line, and z
# values lie *between* 0 and 15 between those lines.
```

```python
x3, y3 = np.meshgrid(np.linspace(-10,10,201),np.linspace(-10,10,201)
)
z3 = np.cos(x3+y3) + 0.05*(x3-y3)
#filled contour plot
plt.contourf(x3,y3,z3)
plt.colorbar()
```
